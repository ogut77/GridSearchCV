```
missing_val_count_by_column = (dt.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
```

```
dt.fillna(dt.mean(), inplace=True)
# count the number of NaN values in each column
print(dt.isnull().sum())
```

```
dt['Target']=np.where(dt['Credit-Application-Result']=='Creditworthy',1,0)
```

```
X=dt
y=X['Target']
X=X.drop(['Credit-Application-Result','Target'],axis=1)
```

```
X.head()
```

```
# Get list of categorical variables
s = (X.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```
y.value_counts()
```

```
drop_X = X.select_dtypes(exclude=['object'])
drop_X.head()
```

```
import statsmodels.api as sm
logit_model=sm.Logit(y,drop_X)
result=logit_model.fit()
print(result.summary2())
```

```
from sklearn.preprocessing import LabelEncoder

# Make copy to avoid changing original data 
label_X = X.copy()

# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
for col in object_cols:
    label_X[col] = label_encoder.fit_transform(X[col])
```

```
label_X.head()
```

```
label_X.dtype()


```

```
label_X.describe()
```

```
del label_X['Concurrent-Credits']
label_X.describe()
```

```
import statsmodels.api as sm
logit_model=sm.Logit(y,label_X)
result=logit_model.fit()
print(result.summary2())
```

```
from sklearn.preprocessing import OneHotEncoder

# Apply one-hot encoder to each column with categorical data
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH = pd.DataFrame(OH_encoder.fit_transform(X[object_cols]))


# One-hot encoding removed index; put it back
OH.index = X.index


# Remove categorical columns (will replace with one-hot encoding)
num_X_train = X.drop(object_cols, axis=1)

# Add one-hot encoded columns to numerical features
OH = pd.concat([num_X_train, OH], axis=1)
import statsmodels.api as sm
logit_model=sm.Logit(y,OH)
result=logit_model.fit()
print(result.summary2())
```
